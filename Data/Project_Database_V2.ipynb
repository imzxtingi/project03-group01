```python
import pandas as pd
import sqlite3
import io

# Assuming you have already created df1, df2, and df3

# Print the DataFrames
print("DataFrame 1:")
print(df1.head())

print("\nDataFrame 2:")
print(df2.head())

print("DataFrame 3:")
print(df3.head())

# Initialize SQLite database in memory
conn = sqlite3.connect(':memory:')

# Store DataFrames in the database
df1.to_sql('table1', conn, if_exists='replace', index=False)
df2.to_sql('table2', conn, if_exists='replace', index=False)
df3.to_sql('table3', conn, if_exists='replace', index=False)

# Create a file-like object to store the database
buffer = io.BytesIO()

# Dump the database as UTF-8 encoded bytes
for line in conn.iterdump():
    buffer.write((line + '\n').encode('utf-8'))

# Close the connection to the in-memory database
conn.close()

# Reset the buffer position to the beginning
buffer.seek(0)

# Write the database to a file
with open('my_database.db', 'wb') as f:
    f.write(buffer.getvalue())

print("DataFrames stored in SQLite database with UTF-8 encoding successfully.")
```
